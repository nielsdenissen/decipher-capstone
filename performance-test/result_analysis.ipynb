```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nlresults = pd.read_csv("results_nl_10e11.csv")
enresults = pd.read_csv("results_en_1e3_2.csv")
deresults = pd.read_csv("results_de_10e11.csv")
```

```python
def plot_result(results):
    results_avg = results.groupby("length").mean()
    
    fig, ax = plt.subplots(1,2, figsize=(15,5))
    ax[0].plot(results_avg.index, results_avg.perc_correct, color='green')
    ax[0].set_title("% Correct")
    ax[0].set_ylim(0,100)
    ax[0].set_xlabel("# words in sentence")
    ax[0].set_ylabel("# words correct")

    ax[1].plot(results_avg.index, results_avg.time)
    ax[1].set_title("Time")
    ax[1].set_xlabel("# words in sentence")
    ax[1].set_ylabel("time (s)")
    fig.show()
```

```python
plot_result(nlresults)
```

```python
plot_result(enresults)
```

```python
plot_result(deresults)
```
